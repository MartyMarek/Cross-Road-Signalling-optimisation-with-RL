```python
log_dir = "final_simulation\\_models\\static"
os.makedirs(log_dir, exist_ok=True)

# Sim time 1 minute
max_simulation_seconds = 180
```

```python
# Define simulation
simulation = SumoSimulation(
    sumo_binary_path="C:\\Program Files (x86)\\Eclipse\\Sumo\\bin\\sumo",
    #sumo_binary_path="C:\\Program Files (x86)\\Eclipse\\Sumo\\bin\\sumo-gui",
    sumo_config_path="C:\\sumoconfig\\real_intersection.sumocfg",
    signal_states=SignalStates
)

# Define environment
env = RealIntersectionStatic(
    simulation=simulation,
    max_simulation_seconds=max_simulation_seconds
)

# Logs will be saved in log_dir/monitor.csv
#env = Monitor(env, log_dir)
```

```python
import pandas as pd

info_df = pd.DataFrame()
n_episodes = 3

for episode in range(1,n_episodes + 1):

    print("Episode: ", episode)

    observation = env.reset()
    step = 0
    done = False

    while done != True:

        step += 1
        observation, reward, done, info = env.step(0)

        data_store = info['traffic'].sum(axis=0).to_dict()
        for key, value in data_store.items():
            data_store[key] = [value]
        data_store['episode'] = [episode]
        data_store['step'] = [step]
        step_traffic_df = pd.DataFrame(data=data_store)
        info_df = pd.concat(
            [info_df,step_traffic_df],
            ignore_index=True
        )

```
